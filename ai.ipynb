9. Implement resolution in propositional logic using Python. 

```python
def resolve(ci, cj):
    """Try to resolve two clauses. Return the resolvent or None."""
    resolvents = []
    for di in ci:
        for dj in cj:
            if di == f"¬¨{dj}" or dj == f"¬¨{di}":
                new_clause = set(ci).union(set(cj)) - {di, dj}
                resolvents.append(tuple(sorted(new_clause)))
    return resolvents

def resolution(kb, query):
    """Apply resolution to check if KB entails the query."""
    clauses = [tuple(sorted(clause)) for clause in kb]
    negated_query = tuple(sorted([f"¬¨{q}" if not q.startswith("¬¨") else q[1:] for q in query]))
    clauses.append(negated_query)

    new = set()
    while True:
        pairs = [(clauses[i], clauses[j]) for i in range(len(clauses)) for j in range(i+1, len(clauses))]
        for (ci, cj) in pairs:
            resolvents = resolve(ci, cj)
            for r in resolvents:
                if not r:  # Empty clause found
                    return True
                new.add(r)
        if new.issubset(set(clauses)):
            return False
        for clause in new:
            if clause not in clauses:
                clauses.append(clause)

# Example Knowledge Base (KB) in CNF
kb = [
    ("A", "B"),
    ("¬¨A",),
    ("¬¨B", "C")
]

# Query: Is C entailed?
query = ("A",)

# Run resolution
result = resolution(kb, query)
print("‚úÖ C is entailed by KB" if result else "‚ùå C is not entailed by KB")
```

10. Create a Forward Chaining rule-based inference system. 

```python
# Initial facts
facts = {"Sunny", "Weekend"}

# Rules: Each rule has a condition (set of facts) and a conclusion
rules = [
    ({"Sunny"}, "GoForWalk"),
    ({"Weekend"}, "Relax"),
    ({"Sunny", "Weekend"}, "Picnic"),
    ({"Picnic"}, "PackFood")
]

# Inferred facts
inferred = set()

def forward_chain(facts, rules):
    new_inferred = True
    while new_inferred:
        new_inferred = False
        for condition, conclusion in rules:
            if condition.issubset(facts) and conclusion not in facts:
                print(f"‚úÖ Rule fired: {condition} ‚Üí {conclusion}")
                facts.add(conclusion)
                inferred.add(conclusion)
                new_inferred = True
    return facts

# Run inference
final_facts = forward_chain(facts.copy(), rules)

# Output results
print("\nüìö Final Facts:")
for fact in final_facts:
    print("-", fact)
```

11. Create a Backward Chaining algorithm using recursion.

```python
# Knowledge base: rules and known facts
rules = {
    "Picnic": [["Sunny", "Weekend"]],
    "GoForWalk": [["Sunny"]],
    "Relax": [["Weekend"]],
    "PackFood": [["Picnic"]],
    "Sunny": [],  # Known fact
    "Weekend": []  # Known fact
}

# Recursive backward chaining function
def backward_chain(goal, visited=None):
    if visited is None:
        visited = set()

    if goal in visited:
        return False  # Prevent infinite loops
    visited.add(goal)

    # If goal is a known fact (no conditions), it's true
    if goal in rules and not rules[goal]:
        print(f"‚úÖ Known fact: {goal}")
        return True

    # Try to satisfy any rule that concludes the goal
    if goal in rules:
        for condition_set in rules[goal]:
            print(f"üîç Trying to prove: {goal} via {condition_set}")
            if all(backward_chain(subgoal, visited.copy()) for subgoal in condition_set):
                print(f"‚úÖ Proven: {goal}")
                return True

    print(f"‚ùå Cannot prove: {goal}")
    return False

# Test the system
goal = "PackFood"
print(f"\nüéØ Goal: {goal}")
result = backward_chain(goal)
print("\nüßæ Result:", "Success" if result else "Failure")
```

12. Simulate Bayesian inference using Python (conditional probabilities). 

```python
def bayesian_inference(prior, likelihood, false_positive):
    # P(E) = P(E|H) * P(H) + P(E|¬¨H) * P(¬¨H)
    marginal = likelihood * prior + false_positive * (1 - prior)
    posterior = (likelihood * prior) / marginal
    return posterior

# Inputs
P_Disease = 0.01
P_Positive_given_Disease = 0.99
P_Positive_given_NoDisease = 0.05

# Compute posterior
P_Disease_given_Positive = bayesian_inference(
    prior=P_Disease,
    likelihood=P_Positive_given_Disease,
    false_positive=P_Positive_given_NoDisease
)

print(f"üßæ Probability of having disease given a positive test: {P_Disease_given_Positive:.4f}")
```

```python
def fuzzify_temperature(temp):
    """Convert crisp temperature to fuzzy categories with degrees."""
    fuzzy = {
        "Cold": max(0, min(1, (25 - temp) / 10)),
        "Warm": max(0, min(1, (temp - 20) / 10 if temp <= 30 else (40 - temp) / 10)),
        "Hot": max(0, min(1, (temp - 30) / 10))
    }
    return fuzzy

def infer_fan_speed(fuzzy_temp):
    """Apply fuzzy rules to determine fan speed."""
    # Rule weights
    low = fuzzy_temp["Cold"]
    medium = fuzzy_temp["Warm"]
    high = fuzzy_temp["Hot"]

    # Defuzzification using weighted average
    speed = (low * 30 + medium * 60 + high * 100) / (low + medium + high)
    return round(speed, 2)

# üîß Test the system
for temp in [18, 22, 27, 33, 38]:
    fuzzy = fuzzify_temperature(temp)
    speed = infer_fan_speed(fuzzy)
    print(f"üå°Ô∏è Temp: {temp}¬∞C ‚Üí Fuzzy: {fuzzy} ‚Üí Fan Speed: {speed}%")
```

Implement a state space problem (e.g., Water Jug Problem) in Python. 
(used BFS)


```python
from collections import deque

CAPACITY_A = 4
CAPACITY_B = 3
GOAL = 2

def water_jug_bfs():
    queue = deque()
    start = (0, 0)
    queue.append((start, [start]))
    visited = set()

    while queue:
        (a, b), path = queue.popleft()

        # Goal check
        if a == GOAL or b == GOAL:
            return path

        # Skip if already visited
        if (a, b) in visited:
            continue

        # Mark current state as visited
        visited.add((a, b))

        # Generate next possible states
        next_states = []
        next_states.append((CAPACITY_A, b))       # Fill Jug A
        next_states.append((a, CAPACITY_B))       # Fill Jug B
        next_states.append((0, b))                # Empty Jug A
        next_states.append((a, 0))                # Empty Jug B

        # Pour A ‚Üí B
        transfer = min(a, CAPACITY_B - b)
        next_states.append((a - transfer, b + transfer))

        # Pour B ‚Üí A
        transfer = min(b, CAPACITY_A - a)
        next_states.append((a + transfer, b - transfer))

        # Add new states to queue
        for state in next_states:
            if state not in visited:
                queue.append((state, path + [state]))

    return None


# Run the BFS
solution = water_jug_bfs()
if solution:
    print("Solution found in", len(solution) - 1, "steps:")
    for step in solution:
        print(f"Jug A: {step[0]} L, Jug B: {step[1]} L")
else:
    print("No solution found.")
```

: Solve the N-Queens problem using Hill Climbing search.

```python
import random

def print_board(state):
    n = len(state)
    for row in range(n):
        line = ""
        for col in range(n):
            if state[col] == row:
                line += "Q "
            else:
                line += ". "
        print(line)
    print("\n")

def calculate_conflicts(state):
    conflicts = 0
    n = len(state)
    for i in range(n):
        for j in range(i+1, n):
            # Same row conflict
            if state[i] == state[j]:
                conflicts += 1
            # Diagonal conflict
            if abs(state[i] - state[j]) == abs(i - j):
                conflicts += 1
    return conflicts

def get_best_neighbor(state):
    n = len(state)
    current_conflicts = calculate_conflicts(state)
    best_state = state[:]
    best_conflicts = current_conflicts

    for col in range(n):
        original_row = state[col]
        for row in range(n):
            if row != original_row:
                state[col] = row
                conflicts = calculate_conflicts(state)
                if conflicts < best_conflicts:
                    best_conflicts = conflicts
                    best_state = state[:]
        # Reset column back to original row after testing
        state[col] = original_row

    return best_state, best_conflicts

def hill_climbing(n):
    # Random initial state
    current_state = [random.randint(0, n-1) for _ in range(n)]
    current_conflicts = calculate_conflicts(current_state)
    steps = 0

    while True:
        if current_conflicts == 0:
            print(f"Solution found in {steps} steps!")
            print_board(current_state)
            return current_state

        neighbor, neighbor_conflicts = get_best_neighbor(current_state)

        if neighbor_conflicts >= current_conflicts:
            print("Reached local minimum (no solution found).")
            print(f"Steps taken: {steps}")
            print_board(current_state)
            return None

        current_state = neighbor
        current_conflicts = neighbor_conflicts
        steps += 1

# Run for 8-Queens
hill_climbing(8)
```

 Implement Best-First Search (Greedy BFS) for pathfinding

```python
import heapq

def best_first_search(start, goal, graph, heuristic):
    frontier = [(heuristic[start], start)]
    came_from = {start: None}
    visited = set([start])

    while frontier:
        _, current_node = heapq.heappop(frontier)

        if current_node == goal:
            break

        for neighbor in graph[current_node]:
            if neighbor not in visited:
                visited.add(neighbor)
                heapq.heappush(frontier, (heuristic[neighbor], neighbor))
                came_from[neighbor] = current_node

    # Build path after search completes
    path = []
    node = goal
    while node is not None:
        path.append(node)
        node = came_from.get(node)
    path.reverse()
    return path


# Graph definition
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G', 'I'],
    'E': ['H'],
    'F': [],
    'G': [],
    'H': [],
    'I': []
}

# Heuristic values
heuristic = {
    'A': 10, 'B': 8, 'C': 7, 'D': 6,
    'E': 4, 'F': 5, 'G': 3, 'H': 2, 'I': 0
}

# Run search
start_node = 'A'
goal_node = 'I'
path = best_first_search(start_node, goal_node, graph, heuristic)

print("Path from {} to {}: {}".format(start_node, goal_node, path))
```

Implement a simple forward chaining inference engine.

```python
def forward_chaining(facts, rules):
    inferred = set(facts)
    changed = True

    print("Initial Facts:", inferred)
    print("-" * 40)

    while changed:
        changed = False

        for conditions, conclusion in rules:
            # Check if all conditions are satisfied
            if set(conditions).issubset(inferred):
                if conclusion not in inferred:
                    inferred.add(conclusion)
                    changed = True
                    print(f"Inferred '{conclusion}' from {conditions}")

    print("-" * 40)
    return inferred


# Initial facts
facts = ["rain", "cold"]

# Rules: (conditions, conclusion)
rules = [
    (["rain", "cold"], "wet"),
    (["wet"], "slippery"),
    (["cold"], "wear_jacket"),
    (["rain"], "carry_umbrella")
]

# Run inference engine
final_facts = forward_chaining(facts, rules)

print("Final Inferred Facts:", final_facts)
```

Write a recursive backward chaining function that checks if a goal fact can be proved from a knowledge base of facts and rules.

```python
def backward_chaining(goal, facts, rules, visited=None):
    if visited is None:
        visited = set()

    print(f"Checking goal: {goal}")

    # If goal is already a known fact
    if goal in facts:
        print(f"‚úî {goal} is a known fact")
        return True

    # Avoid infinite loops
    if goal in visited:
        return False

    visited.add(goal)

    # Try to prove goal using rules
    for conditions, conclusion in rules:
        if conclusion == goal:
            print(f"Trying rule: IF {conditions} THEN {conclusion}")

            # Check if all conditions can be proved
            if all(backward_chaining(cond, facts, rules, visited) for cond in conditions):
                print(f"‚úî {goal} proven using rule {conditions} ‚Üí {conclusion}")
                return True

    print(f"‚úò Cannot prove {goal}")
    return False


# Knowledge base
facts = {"rain", "cold"}

rules = [
    (["rain", "cold"], "wet"),
    (["wet"], "slippery"),
    (["cold"], "wear_jacket")
]

# Test backward chaining
goal = "wet"
result = backward_chaining(goal, facts, rules)

print("\nFinal Result:")
print(f"Can '{goal}' be proved? ->", result)
```

Write a function to calculate the posterior probability

```python
def calculate_posterior():
    # Given probabilities
    p_disease = 0.01
    p_no_disease = 1 - p_disease

    p_positive_given_disease = 0.9
    p_positive_given_no_disease = 0.05

    # Bayes' theorem
    numerator = p_positive_given_disease * p_disease
    denominator = (p_positive_given_disease * p_disease) + \
                  (p_positive_given_no_disease * p_no_disease)

    posterior = numerator / denominator
    return posterior


# Calculate and display result
result = calculate_posterior()
print("P(Disease | PositiveTest) =", result)
```
